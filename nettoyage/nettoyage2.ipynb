```python
import pandas as pd

def drop_columns(liste_colomuns, df): # df = chemin du fichier csv
    
    dataframe = pd.read_csv(df, sep = ";")
    
    dataframe = dataframe.drop(liste_colomuns, axis = 1)
    return dataframe

def indicateur(df,col_name = "Indicateur" ):
    
    dataframes = {}
    indic_uniques = df[col_name].unique()
    
    for elt in indic_uniques:
        dataframes[elt] = df[df[col_name] == elt]
    return dataframes


formation = pd.read_csv("bilan-social-d-edf-sa-formation.csv", sep = ";")
indicateur = indicateur(formation)
for k, v in indicateur.items():
    print(v)
    
    
def indicateur2(df, col_name = "Indicateur"):
    df = pd.read_csv(df, sep = ";")
    groupements = df.groupby([col_name])
    
    for key, group in groupements:
        group.to_csv(f'groupe_{key}.csv', index = False)

indicateur2("bilan-social-d-edf-sa-remuneration-et-promotions.csv")
```
